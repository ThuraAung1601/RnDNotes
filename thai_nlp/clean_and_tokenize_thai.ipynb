```python
from pythainlp.tokenize import word_tokenize

invisible_chars = [
    r"\xa0",   # Non-breaking space
    r"\u200B",  # Zero-width space
    r"\u200C",  # Zero-width non-joiner
    r"\u200D",  # Zero-width joiner
    r"\u200E",  # Left-to-right mark
    r"\u200F",  # Right-to-left mark
    r"\t",      # Tab
]

def clean(sentence, to_clean_char_list=invisible_chars):
    pattern = "|".join(to_clean_char_list)
    cleaned_sentence = re.sub(pattern, "", sentence)
    return cleaned_sentence
    
data = ""  

with open("test.txt", "r", encoding="utf-8") as file:
    for line in file:
        line = line.strip()
        # print(line)
        if line: 
            tokenized_line = word_tokenize(clean(line), engine="newmm")
            tokenized_line = [token for token in tokenized_line if token.strip()]
            line_with_tokens = "|".join(tokenized_line) + "\n"
            data += line_with_tokens 
            
with open("output.txt", "w", encoding="utf-8") as file:
    file.write(data)  
```

```python
from pythainlp.tokenize import word_tokenize
from pythainlp.util import normalize
    
data = ""  

with open("test.txt", "r", encoding="utf-8") as file:
    for line in file:
        line = line.strip()
        # print(line)
        if line: 
            tokenized_line = word_tokenize(normalize(line), engine="newmm")
            tokenized_line = [token for token in tokenized_line if token.strip()]
            line_with_tokens = "|".join(tokenized_line) + "\n"
            data += line_with_tokens 
            
with open("output.txt", "w", encoding="utf-8") as file:
    file.write(data)  
```
